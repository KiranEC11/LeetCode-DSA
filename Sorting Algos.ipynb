## selection sort:

select minimum and swap

```python
## time complexity O(n2)
```

```python
from typing import List

def selectionSort(arr: List[int]):
    n = len(arr)
    for i in range(n-1):

        min_ind = i
        for j in range(i,n):
            if arr[j]< arr[min_ind]:  ## finding min value index
                min_ind = j
        arr[i], arr[min_ind] = arr[min_ind], arr[i]  ##swap
        # print(arr)

    return arr
```

```python
selectionSort([2, 13, 4, 1 ,3 ,6 ])
```

### Bubble sort

push max to the end

```python
from typing import List

def bubbleSort(arr: List[int], n: int):
    # Your code goes here.
    n = len(arr)
    for i in range(n):  ## each step
        for j in range(n-1-i):  ## hovering over elements
            if arr[j+1]<arr[j]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
```

## insertion sort

iterate over each element, take the list upto that element and place that element in the correct order. ( basically we are sorting each subarrays step by step)

Time complexity

worst case -- O(n2)

avg -- O(n2)

best - **O(n)**

```python
from typing import List

def insertionSort(a: List[int], n: int) -> None:
   # write your code here

   for i in range(1,n):
      j = i
      while j>0 and a[j]<a[j-1]:
         a[j], a[j-1] = a[j-1], a[j]

         j-=1
```

# merge sort

Divide and merge

space -- o(n)

```python
from typing import List

def merge(arr, low, mid, high):
    temp =[]
    left = low
    right = mid+1

    while left<=mid and right<=high:
        if arr[left]<=arr[right]:
            temp.append(arr[left])
            left+=1
        else:
            temp.append(arr[right])
            right+=1
    while left<=mid:
        temp.append(arr[left])
        left+=1
    while right<=high:
        temp.append(arr[right])
        right+=1

    for i in range(low,high+1):
        arr[i] = temp[i-low]

def mergeSort(arr: List[int], low: int, high: int):
    # Write Your Code Here

    def ms(arr, low, high):
        if low==high:
            return


        mid  = (low+high)//2
        ms(arr, low,mid)
        ms(arr, mid+1, high)
        merge(arr, low, mid, high)
        # return arr  --> ### return nothing: uncomment this for returning arr
    return ms(arr, low, high)



## if u want to return arr, then modify accordingly.
```

```python
"""
	The function is called with the parameters:
	quickSort(input, 0, size - 1);

"""
from typing import List

def quickSort(arr: List[int], startIndex: int, endIndex: int):
    """
    Don't write main().
    Don't read input, it is passed as function argument.
    Change in the given array itself.
    Taking input and printing output is handled automatically.
    """

    def f(arr, low, high):
        if low>=high:
            return

        partition_id = partition(arr, low, high)
        f(arr, low, partition_id-1)
        f(arr, partition_id+1, high)

    f(arr, startIndex, endIndex)



def partition(arr, low, high):
    pivot = arr[low]
    l = low
    r = high
    while l<r:
        while l<=high and arr[l]<=pivot:  ## find the index of the first ele which is larger than pivot (from the beginnig- low)

            l+=1

        while r>=low and arr[r]>pivot:   ## find the index of the first ele which is smaller than pivot (from the end- high)

                r-=1

        if l<r:
            arr[l], arr[r] = arr[r], arr[l] ## swap larger and smaller
    arr[r], arr[low] = arr[low], arr[r]  ## swap the pivotted ele and first ele from the end which is smaller than pivot

    return r
```
