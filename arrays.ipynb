Given an m x n integer matrix matrix, if an element is 0, set its entire row and column to 0's.

You must do it in place.



Example 1:


Input: matrix = [[1,1,1],[1,0,1],[1,1,1]]

Output: [[1,0,1],[0,0,0],[1,0,1]]

### important

Follow up:

A straightforward solution using O(mn) space is probably a bad idea.

A simple improvement uses O(m + n) space, but still not the best solution.

Could you devise a constant space solution?  --> YES I HAVE DONE IT

```python
class Solution:
    def setZeroes(self, matrix: list[list[int]]) -> None:
        """
        Do not return anything, modify matrix in-place instead.
        """

        ## Brute force approach
        ## O(m*n*(m+n) + m*n)

        # def markRow(row):
        #     for col in range(len(matrix[0])):
        #         if matrix[row][col]!=0:
        #             matrix[row][col] = -1000 # any negative number. (we assume the entries in matrix are non negative)
        # def markCol(col):
        #     for row in range(len(matrix)):
        #         if matrix[row][col]!=0:
        #             matrix[row][col] = -1000


        # for row in range(len(matrix)):
        #     for col in range(len(matrix[0])):
        #         if matrix[row][col]==0:
        #             markRow(row)
        #             markCol(col)

        # for row in range(len(matrix)):
        #     for col in range(len(matrix[0])):
        #         if matrix[row][col]==-1000:
        #             matrix[row][col]=0


        ############################################################

        ## O(3*mn)
        # def markRow(rows):
        #     for row in rows:
        #         for col in range(len(matrix[0])):
        #                 matrix[row][col] = 0 # any negative number. (we assume the entries in matrix are non negative)
        # def markCol(cols):
        #     for col in cols:
        #         for row in range(len(matrix)):
        #             matrix[row][col] = 0

        # rows = []
        # cols = []
        # for row in range(len(matrix)):
        #     for col in range(len(matrix[0])):
        #         if matrix[row][col]==0:
        #             # markRow(row)
        #             # markCol(col)
        #             rows.append(row)
        #             cols.append(col)
        # markRow(rows)
        # markCol(cols)

######################################

### O(2mn) time complexity

### O(m+n)  space complexity

        # n = len(matrix)
        # m=len(matrix[0])

        # rows = [0]*n
        # cols = [0]*m
        # for row in range(n):
        #     for col in range(m):
        #         if matrix[row][col]==0:
        #             # markRow(row)
        #             # markCol(col)
        #             # rows.append(row)
        #             # cols.append(col)
        #             rows[row]=1
        #             cols[col]=1
        # for row in range(n):
        #     for col in range(m):
        #         if matrix[row][col]!=0:
        #             if rows[row] ==1 or cols[col]==1:
        #                 matrix[row][col] =0

################################

       ### time compexity cannot be reduced
       ### reduce space complexity to O(1)

       ## take rows and cols in above code as the first row and col of matrix itself.
       ## warning: to avoid taking matrix[0][0] in both the array, take it only in rows.
       ## declare a new variable col0 for col 0 reference. so len(cols) will be m-1,len(rows)=n

        col0=1
        n=len(matrix)
        m= len(matrix[0])

        for i in range(n):
            for j in range(m):
               if matrix[i][j] ==0:
                    if j!=0:
                        matrix[i][0]=0
                        matrix[0][j] =0
                    else:
                        matrix[i][0] =0
                        col0=0
        for i in range(1,n):
            for j in range(1,m):
                if matrix[i][j]!=0:
                    if matrix[i][0]==0 or matrix[0][j]==0:
                        matrix[i][j]=0

        ## the order of update in below is important.
        ## fist row{1 to n} is not reference to changing any ele in first col(it depends only on col0). so we can change that first.

        if matrix[0][0]==0:
            for j in range(1,m):
                matrix[0][j]=0
        if col0==0:
            for i in range(n):
                matrix[i][0]=0
```
